## Loan Dataset : Data Cleaning

```python
# Finding missing value for each column.
loan_data.isnull().sum()
```

```python
# Finding missing value (returns boolean) for each column. If there's at least one missing value in a column, result is true.
loan_data.isnull().any()
```

```python
# Finding percentage of missing values
miss_percent = round(100*(loan_data.isnull().sum()/len(loan_data.index)),2)
miss_percent[miss_percent > 0]
```

```python
# Dropping columns more than 25 percent missing or null values
loan_data.drop(miss_percent[miss_percent >25].index, axis = 1, inplace = True)
```

```python
# Provides the total number of rows and columns after dropping columns with >25% null values
loan_data.shape
```

```python
# Finding the number of unique values in each column
loan_data.nunique().sort_values()
```

```python
# Dropping columns with only one unique value
loan_data.drop(['tax_liens','delinq_amnt','chargeoff_within_12_mths','acc_now_delinq','application_type','policy_code','collections_12_mths_ex_med','initial_list_status','pymnt_plan'], axis = 1, inplace = True)
```

```python
# Provides the total number of rows and columns after dropping columns with one unique value
loan_data.shape
```

```python
# Viewing the first two rows of the dataset
loan_data.head(2)
```

```python
# Either id or member_id can be used can primary key. Hence, we can drop one of them ie.member_id 
loan_data.drop(['member_id'], axis = 1, inplace = True)
```

```python
# Dropping url column
loan_data.drop(['url'], axis = 1, inplace = True)
```

```python
#Checking again the rows and columns after dropping 
loan_data.shape
```

```python
# Viewing all the columns
pd.set_option('max_columns', None)
loan_data.head()
```

```python
# Returns the data type of each column
loan_data.dtypes
```

```python
# Stripping % from int_rate and revol_util and convert to float
loan_data['int_rate'] = loan_data.int_rate.str.strip('%').astype(float)
loan_data['revol_util'] = loan_data.revol_util.str.strip('%').astype(float)
```

```python
# Viewing all the columns
pd.set_option('max_columns', None)
loan_data.head()
```

```python
# Value count for Loan_status 
loan_data['loan_status'].value_counts()
```

```python
# Removing 'Current' in Loan_status as it will not aid in our analysis. Focus is not Fully Paid and Charged off.
loan_data = loan_data[loan_data.loan_status != 'Current']
```

```python
# Rechecking the removal'Current' in Loan_status
loan_data['loan_status'].value_counts()
```

```python
# Loan_status percentage
loan_stats_percnt = loan_data['loan_status'].value_counts(normalize=True)
loan_stats_percnt
```

```python
# Further dropping emp_title and title as they have many unique text values
loan_data.drop(['emp_title','title'], axis=1, inplace = True)
```

```python
# Returns the first 20 rows of the loan dataset
loan_data.head(20)
```

```python
# Dropping customer behaviour variables total_rec_int,total_rec_prncp,total_rec_late_fee, recoveries, collection_recovery_fee,last_credit_pull_d,last_pymnt_d, out_prncp,out_prncp_inv
loan_data.drop(['total_rec_int','total_rec_prncp','total_rec_late_fee','last_credit_pull_d','last_pymnt_d', 'out_prncp','out_prncp_inv','recoveries','collection_recovery_fee'], axis = 1, inplace = True)
```

```python
# Viewing the rows and colums
loan_data.shape
```

```python
# Finding the missing values
miss_percent = round(100*(loan_data.isnull().sum()/len(loan_data.index)),2)
miss_percent[miss_percent > 0]
```

```python
# Returns the bolean of missing values in the dataset
loan_data.isnull().any()
```

```python
# Further dropping customer behaviour variables with missing values
loan_data.dropna(subset=['revol_util', 'pub_rec_bankruptcies'], inplace = True)
loan_data.shape
```

```python
# Removing 'months' from term column for easy analysis.
loan_data['term'] = loan_data.term.apply(lambda x: x.split() [0])
loan_data.term.value_counts()
```

```python
# Possible values are between 0 and 10 as per data dictionary provided. Coverting values between 0 to 10
emp_length_newval = {
    '< 1 year' : 0,
    '1 year'   : 1,
    '2 years'  : 2,
    '3 years'  : 3,
    '4 years'  : 4,
    '5 years'  : 5,
    '6 years'  : 6,
    '7 years'  : 7,
    '8 years'  : 8,
    '9 years'  : 9,
    '10+ years': 10
}
# replacing the values in emp_length with emp_length_newval
loan_data = loan_data.replace({"emp_length": emp_length_newval})
loan_data.emp_length.value_counts()
```

```python
# Checking the datatypes
loan_data.dtypes
```

```python
# Converting the dtype for issue_d, earliest_cr_line to datetime
loan_data.issue_d = pd.to_datetime(loan_data.issue_d, format='%b-%y')
loan_data.earliest_cr_line = pd.to_datetime(loan_data.earliest_cr_line, format='%b-%y')
```

```python
# Spliting issue_d and earliest_cr_line into month and year columns
loan_data['issue_d_month'] = loan_data['issue_d'].dt.month
loan_data['issue_d_year'] = loan_data['issue_d'].dt.year
loan_data['earliest_cr_line_month'] = loan_data['earliest_cr_line'].dt.month
loan_data['earliest_cr_line_year'] = loan_data['earliest_cr_line'].dt.year
```

```python
# Under loan_status convert fully paid to 0 and charged off to 1, easier while making analysis.
loan_data["loan_status"] = loan_data["loan_status"].apply(lambda x: 0 if x =="Fully Paid" else 1)
loan_data["loan_status"]
```

```python
# returns information on the loan_data dataset
loan_data.info()
```

```python
# View the amount of defaulters in loan_status - fully paid(0) vs Charged off(1) using count plot
fig = plt.figure(figsize=(3,5))
sns.countplot(loan_data['loan_status'],palette=['#432371',"#FAAE7B"])
plt.show()
```

```python
# Creating bins for loan_amnt_range
bins = [0,5000,10000,15000,20000,25000,36000]
bucket = ['0-5000','5000-10000','10000-15000','15000-20000','20000-25000','25000+']
loan_data['loan_amnt_range'] = pd.cut(loan_data['loan_amnt'],bins,labels=bucket)
loan_data['loan_amnt_range']
```

```python
# Creating bins for int_rate_range
bins = [0,7.5,10,12.5,15,100]
bucket = ['0-7.5','7.5-10','10-12.5','12.5-15','15+']
loan_data['int_rate_range'] = pd.cut(loan_data['int_rate'],bins,labels=bucket)
loan_data['int_rate_range']
```

```python
# Creating bins for dti_range
bins = [-1,5.00,10.00,15.00,20.00,25.00,50.00]
bucket = ['0-5%','5-10%','10-15%','15-20%','20-25%','25%+']
loan_data['dti_range'] = pd.cut(loan_data['dti'],bins,labels=bucket)
loan_data['dti_range']
```

```python
# Creating bins for annual_inc range
bins = [0,25000,50000,75000,100000,1000000]
bucket = ['0-25000','25000-50000','50000-75000','75000-100000','100000+']
loan_data['annual_inc_range'] = pd.cut(loan_data['annual_inc'],bins,labels=bucket)
loan_data['annual_inc_range']
```

```python
# Function to plot univariate with default loan_status scale 0 (fully paid) - 1(charged off)
def plotUnivariateBar(x, figsize=(10,5)):
    plt.figure(figsize=figsize)
    sns.barplot(x=x, y='loan_status', data=loan_data)
    plt.title("Loan Defaults wrt to "+str(x)+" feature using countplot")
    plt.xlabel(x, fontsize=16)
    plt.ylabel("defaulter ratio", fontsize=16)
    plt.show()
```

```python
# Bar plot to view the defaulters ratio by loan term - 36 months and 60 months
plotUnivariateBar("term",figsize=(3,5))
```

#### Observation for defaulters vs term : Increase in defaulters for loans with 60months tenure

```python
# Function to plot boxplot for comparing relationship between two variables
def boxplot_1(x,y, hue="loan_status"):
    plt.figure(figsize=(12,10))
    sns.boxplot(x=x,y=y, data = loan_data, hue=hue, order=sorted(loan_data[x].unique()))
    plt.title("box plot "+x+" vs "+y+" for "+hue)
    plt.xlabel(x, fontsize=15)
    plt.ylabel(y, fontsize=15)
    plt.show()
```

```python
# Plot boxplot for term vs int_rate for loan_status
boxplot_1("term","int_rate")
```

### Observation: High defaulters for higher int_rate and 60 months term

```python
# Plot boxplot for grade vs int_rate for loan_status
boxplot_1("grade","int_rate")
```

### Observation : It's observed there's a gradual increase in defaulters as the grade and int_rate increases

```python
# Function to plot scatter plots for comparing relationship between two variables
def scatter_1(x,y):
    plt.figure(figsize=(12,6))
    sns.scatterplot(x=x,y=y, hue="loan_status", data = loan_data)
    plt.title("Scatter plot between "+x+" and "+y)
    plt.xlabel(x, fontsize=16)
    plt.ylabel(y, fontsize=16)
    plt.show()
```

```python
# View scatter plot between annual_inc and loan_amnt w.r.t loan_status
scatter_1("annual_inc", "loan_amnt")
```

## Multivariate Analysis using Heatmaps - using Seaborn

```python
# Plot heatmap to see correlation between variables in loan data set
plt.figure(figsize=(12,9))
var1 = ["loan_amnt","annual_inc","term","int_rate","grade","sub_grade","dti","loan_status","addr_state"]
loan_correlation = loan_data[var1].corr()
sns.heatmap(loan_correlation, vmin=-1.0, vmax=1.0,annot=True,cmap="Blues")
plt.title("Correlation Heatmap")
plt.show()
```

### Observation : Strong correlation is observed between variables int_rate, annual_inc and dti
